# **Import the necessary libraries**

```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
import seaborn as sns

#Set the matplotlib backend to the 'inline' backend
%matplotlib inline
```

# **Load the *iris* Database**

```
from sklearn import datasets
iris_dataset = datasets.load_iris()
```

# **Explore the *iris* Database**

### **Display the shape of the dataset**

```
#The shape of the data array : (the number of samples , the number of features).
print("shape of data: {}".format(iris_dataset['data'].shape))
```

### **Display the classes (target) in *iris* dataset**

```
print("Target: \n{}".format(iris_dataset['target']))
```

###**Visualize your database**

```
iris_dataset
```

*   **Target** : the desired output
*   **Number of instances**: 150 (50 in each of three classes)
*   **3 classes**: Iris-Setosa, Iris-Versicolour, Iris-Virginica
*   **4 features**: sepal length, sepal width, petal length, petal width 

# **Separate training / testing set using *train_test_split***

```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(iris_dataset['data'],iris_dataset['target'], random_state=0, train_size=0.80)
```

### **Display the target of train set y_train**

```
print("X_train shape: {}".format(X_train.shape))
print("y_train shape: {}".format(y_train.shape))
y_train
```

### **Display the target of test set y_test**

```
print("X_test shape: {}".format(X_test.shape))
print("y_test shape: {}".format(y_test.shape))
y_test
```

# **Create your classification model** : Import the knn classifier (**KNeighborsClassifier**) from scikit-learn and create your model

```
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
```

# **Model Training : Train the knn model from the training set using the method *fit***

```
knn.fit(X_train,y_train)
```

# **Make Predictions using the method *predict***

### **Prediction of a new sample**

```
#Create a new sample
x_new =np.array([[5, 2.9, 1, 0.2]])
print("x_new shape : {}".format(x_new.shape))
```

```
#Predict the class of the new sample; the new flower
prediction = knn.predict(x_new)
print("Prediction: {}".format(prediction))
print("predicted target name: {}".format(iris_dataset['target_names'][prediction]))
```

### **Prediction of the Test set X_test**

```
y_pred = knn.predict(X_test)
print("Test set predictions: \n {}".format(y_pred))
```

# **Measure the performance of the predictor : Test Accuracy**

**Test the accuracy of the model using the score function**

```
print("Test set accuracy: {:2f}".format(knn.score(X_test, y_test)))
error = 1 - knn.score(X_test, y_test)
print("Erreur: %f" % error)
```
